```python
df = df.dropna()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['sales_date'] = pd.to_datetime(df['sales_date'])
```

```python
numeric_cols = ['sales_amount', 'quantity_sold', 'unit_cost', 'unit_price', 'discount']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```python
df['total_cost'] = df['unit_cost'] * df['quantity_sold']
```

```python
df['profit'] = df['sales_amount'] - df['total_cost']
```

```python
df['profit_margin'] = df['profit'] / df['sales_amount']
```

```python
text_cols = ['region', 'sales_rep', 'product_category', 'payment_method', 'sales_channel']
df[text_cols] = df[text_cols].apply(lambda x: x.str.strip().str.title())
```

```python
df.to_csv("cleaned_sales_data.csv", index=False)
```
